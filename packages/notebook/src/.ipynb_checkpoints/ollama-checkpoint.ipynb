```typescript
import { ChatOllama } from "npm:@langchain/ollama";

const model = new ChatOllama({
  model: "deepcoder:14b",
});
```

```typescript
const modelTestResult = await model.stream("hello? who are you?")

const chunks = [];
for await (const chunk of modelTestResult) {
  chunks.push(chunk);
  console.log(`${chunk.content}`);
}
console.log(chunks);
```

```typescript
const codingResult = await model.stream(CODE_PROMPT)

const chunks = [];
for await (const chunk of codingResult) {
  chunks.push(chunk);
  console.log(`${chunk.content}`);
}

let finalChunk = "";

for (const chunk of chunks) {
  finalChunk = finalChunk.concat(chunk);
}

finalChunk;
```
